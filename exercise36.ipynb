# Line Plot

```python
x = [1, 2, 3, 4, 5, 6]
y = [200, 110, 70, 10, 0, 30]

plt.plot(x, y)
plt.show()
```

# Bar plot

```python
from matplotlib import pyplot as pyplot

x = [10, 20, 30, 45, 2]
y = [30, 20, 100, 200, 50]

plt.bar(x, y)
plt.show()
```

# Scatter Plot

```python
from matplotlib import pyplot as plt

x = [5, 1, 3, 4, 7]
y = [10, 20, 2, 10, 5]

plt.scatter(x, y)
plt.show()
```

## Adding information

```python
from matplotlib import pyplot as plt

x = [5, 1, 3, 4, 7]
y = [10, 20, 2, 10, 5]

plt.title("Mga nag miss")
plt.xlabel("Number of people")
plt.ylabel("Mga target")

plt.scatter(x, y)
plt.show()
```

## Combining graphs

```python
from matplotlib import pyplot as plt

x = [5, 1, 3, 4, 7]
y = [10, 20, 2, 10, 5]

plt.scatter(x, y)
plt.plot(x, y)
plt.show()

plt.bar(x, y)
plt.scatter(x, y)
plt.show()
```
