```python
import pandas as pd
from sklearn.model_selection import train_test_split

print(pd.__version__)
data = pd.read_csv("titanic3.csv")

for column in ["pclass", "age", "fare"]:
    data[column] = (data[column]
        .fillna(
            data[column].mean().astype(int)
        )
    )

data["sex"] = data["sex"].replace({
    "male": 1,
    "female": 0,
})
```

# Split the data into training data and test data

```python
x_train, x_test, y_train, y_test = train_test_split(
    data[["sex", "pclass", "age", "fare"]],
    data.survived,
    test_size=0.2,
    random_state=0,
)

print(x_train[:3])
print(y_train[:3])
```

# Pre-processing of data
Here, we normalize the features into the range [0..1].
For example, instead of age 53, it'll be assigned as .53

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# Train Naive Bayes model

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(x_train, y_train)
```

# Attempt to predict
input: ["sex", "pclass", "age", "fare"]

sex: "male": 1, "female": 0,

```python
to_predict = [
    {"sex": 1, "pclass": 2, "age": 7, "fare": 120}
]

to_predict = pd.DataFrame(to_predict)

to_predict = sc.transform(to_predict)
print(model.predict(to_predict))
```

# Determine accuracy of model

```python
from sklearn import metrics

predict_test = model.predict(x_test)
print(metrics.accuracy_score(y_test, predict_test))
```
